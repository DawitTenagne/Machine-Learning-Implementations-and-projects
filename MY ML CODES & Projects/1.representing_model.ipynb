````python
#1
import numpy as np 
import matplotlib.pyplot as plt


#2 features
x_train = np.array([1,2])
y_train = np.array([300,500])

#3 the number of training examples

m = x_train.shape[0] #shape will give a tuple = (2,), so i use index to extract the 1st value

#4 printing the training example using for loop

for i in range(m): #range(2)= 0,1 
    print(f"{i}th training example = ({x_train[i]},{y_train[i]})")

#5 ploting the data using scatter plot
"""
plt.scatter(x_train, y_train) #just simple plot, style will be the default
plt.xlabel("features / inputs")
plt.ylabel("targets/ outputs")"""

#6 creating a function that compute the value of the feature using x,w,b as input

#f_wb = wx+b....this is the formula of linear regression

def output_computer(w,x,b):
    # x = input values array,  w and b = parameters that will be adjusted
    #let the arguments are x = [1,2,3,4,5],  w=100, b=100
    # n = #the number of elements in the array / number of inputs
    
    n = x.shape[0] # n = 5
    f_wb = np.zeros(n) # = [0,0,0,0,0] creating a zeros array for the output so that it will be easier to change teh elements later, instead of ammending every time
    for i in range(n): # i = 0,1,2,3,4
        f_wb[i] = (w * x[i]) + b
    return f_wb

#7 & #8 tesing my model with different values of w and b, and comparing the plotting of the actual data and the predicted data

# ploting the actual data from the training examples at 1st using just scatter points

plt.scatter(x_train,y_train, color="red", label="actual data")


# checking my function prediction with diferent values of w and b and ploting it using line plot

w = 100
b = 200
x = x_train
y = output_computer(w,x,b)

plt.plot(x,y, color="blue",marker="x", label="prediction")
    

````
